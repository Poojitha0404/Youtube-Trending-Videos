#### Handling missing values

```python
# category fill
df['category'] = df['category'].fillna(df['category'].mode().iloc[0])

# numeric fill with median
for col in ['likes', 'comment_count']:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

#### Detect & Handle outliers(views) - two methods

```python
## Outliers detection
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(num_cols)
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {outlier_count} outliers")
```

```python
Q1 = df['views'].quantile(0.25)
Q3 = df['views'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df['views_outlier'] = ((df['views'] < lower) | (df['views'] > upper))
df['views_outlier'].sum()
```

```python
df_no_outliers = df[(df['views'] >= lower) & (df['views'] <= upper)]
```

### Text analysis (title/tags) - quick start

```python
# Most common words in titles (simple version)
from collections import Counter
import re

all_titles = " ".join(df['title'].astype(str)).lower()
words = re.findall(r'\w+', all_titles)
stopwords = set(['video','title','new','official','2023','ft','feat'])
common = Counter(w for w in words if w not in stopwords)
common.most_common(30)
```
